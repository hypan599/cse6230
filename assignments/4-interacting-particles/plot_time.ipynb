```python
import matplotlib.pyplot as plt
```

```python
x = [2 ** i for i in range(4, 10)]
" ".join([str(i) for i in x])
```

```python
nt_cloud = [80000, 40000, 16000, 8000, 2500, 710]
" ".join([str(i) for i in nt_cloud])
```

```python
nt_cloud2 = [120000, 30000, 8000, 2000, 600, 200]
```

```python
stream_time = [0.413, 0.293,0.15,0.06,0.04,0]
```

```python
accelerate_time = [0.533, 0.857,0.713,0.815,0.958,0.910]
```

```python
accelerate2_time = [2.68, 2.15, 1.64, 1.1, 0.88, 0.688]
```

```python
stream_y = [10 ** 6 * stream_time[i] / x[i] / nt_cloud[i] for i in range(len(x))]
accelerate_y = [10 ** 6 * accelerate_time[i] / x[i] / nt_cloud[i] for i in range(len(x))]
accelerate2_y = [10 ** 6 * accelerate2_time[i] / x[i] / nt_cloud2[i] for i in range(len(x))]
```

```python
fig = plt.figure(figsize=(20, 10))

plt.plot(x, stream_y, label="stream and noise (cloud)")
plt.plot(x, accelerate_y, label="accelerate")
plt.plot(x, accelerate2_y, label="accelerate2")

plt.rcParams["font.size"] = 20
plt.legend()
plt.ylabel("walltime per particle time step (/1e-6s)")
plt.xlabel("number of particles")

plt.savefig("cloud_walltime_compare.png")
```
