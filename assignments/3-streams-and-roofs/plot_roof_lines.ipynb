```python
import matplotlib.pyplot as plt
```

```python
def roof_line(x, roof, balance):
    if x > balance:
        return roof
    else:
        return roof / balance * x
```

```python
x = list(range(800))
```

```python
# all values in GB
cpu_bp = [roof_line(i, 332, 332/56) for i in x]
gpu_bp = [roof_line(i, 4700, 4700/6) for i in x]
gpu_bp_hbm = [roof_line(i, 4700, 4700/530) for i in x]
```

```python
fig = plt.figure(figsize=(20, 10))

plt.plot(x, cpu_bp, label="CPU")
plt.plot(x, gpu_bp, label="CPU->GPU->CPU")
plt.plot(x, gpu_bp_hbm, label="GPU")

plt.axvline(7/24,  alpha=0.3)

plt.legend()
plt.ylabel("Gflop/s")
plt.xlabel("double precision flops / byte")

plt.savefig("threerooflines.png")
```
