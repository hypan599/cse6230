```python
import math
import matplotlib.pyplot as plt
```

```python
def f(n, p, l = 1e-3, m=1e-6):
    return (n**2 + 2 * m * n)/p + 2 * math.sqrt(p) + 3 * n / math.sqrt(p) + 2 * l * math.log(p)
```

```python
N = 1e5
print(math.log(N))
xs = range(int(math.log(N) * N), int(math.log(N) ** 2 * N), int(math.sqrt(N)))
ys = [f(N, p) for p in xs]
plt.plot(xs, ys)
```
